## A Quick Recap
In Lab 1, Exercise 2 we created a simple actor network in the following way: 

```python
import matplotlib.pyplot as plt
import networkx as nx
actorRels = [
["Marlon Brando", "Al Pacino"],
["Marlon Brando", "Henry Silva"],
["Henry Silva", "Al Pacino"],
["Al Pacino", "Michelle Pfeiffer"]
]
actorGraph = nx.Graph()
actorGraph.add_edges_from(actorRels)
nx.draw_networkx(actorGraph)
```

The display wasn't great, so we adjusted the font size of our labels and the size of our nodes.  

```python
nx.draw_networkx(actorGraph, font_size=7, node_size=10)
plt.show()
```

Now, when it looks better, we can save our visualisation of the network.

## Exercise 1: Saving the visualisation as an image
To save our visualisation of `actorGraph` as an image we will use `pyplot` from `matplotlib` (our `plt`). 
More info: https://matplotlib.org/stable/tutorials/pyplot.html 
We can save it as a PNG file by using `savefig`.

*Remember: We will have to draw our network again with NetworkX and then call the `plt.savefig`*

```python
nx.draw_networkx(actorGraph, font_size=7, node_size=10)
plt.savefig("actorGraph.png", format="PNG")
```

## Exercise 2: Saving information about the network
If we want to store info about our network and/or reuse it in the future, we should probably save it in some other format than PNG. 

One of the formats used in network science is pajek net format.  
Pajek NET is a text-based file format for storing graphs.

To create a pajek net file from NetworkX (our `nx`) we use `write_pajek`: 

```python
nx.write_pajek(actorGraph, "actorGraph.net", encoding='UTF-8')
```

Check whether the is a new file called actorGraph.net in the same folder where you've saved this notebook. 
If so, congratulations on saving your first graph! 

## Exercise 3: Opening saved graphs 
Now let's test whether we can actually read the file we've just created and load all the information we've gathered so far about our network. 

For that we will create a new graph, called `G` and use the NetworkX (our `nx`) function `read_pajek` to load all the info stored in `actorGraph.net` and then we will use the NetworkX function `draw_network` to display our graph.

```python

G = nx.read_pajek("actorGraph.net", encoding='UTF-8')
nx.draw_networkx(G)
```

The display trims the edges again, but we won't care about this for now. 

Congratulations on loading and displaying in your first network! 

## Exercise 4: Adding edges to the network

Now, let's assume we want to add information about a new relationship: 
Al Pacino co-starring with Margot Robbie in *Once Upon a Time... in Hollywood* (2019).
We can use `add_edge` method to do that.  

```python
G.add_edge("Al Pacino", "Margot Robbie")
nx.draw_networkx(G)
```

We can add more than one new relationship by creating a new list of relationships and additing it with `add_edges_from` method.

Let's add three more actors who starred in *Once Upon a Time...in Hollywood* together with Robbie and Pacino: Leonardo DiCaprio, Brad Pitt, and Emile Hirsch.


```python
additionalRels = [
["Margot Robbie", "Leonardo DiCaprio"],
["Margot Robbie", "Brad Pitt"],
["Margot Robbie", "Emile Hirsch"],
["Al Pacino", "Leonardo DiCaprio"],
["Al Pacino", "Brad Pitt"],
["Al Pacino", "Emile Hirsch"],
["Leonardo DiCaprio", "Brad Pitt"],
["Leonardo DiCaprio", "Emile Hirsch"],
["Brad Pitt", "Emile Hirsch"]
]
G.add_edges_from(additionalRels)
nx.draw_networkx(G)
```

# Exercise 5: Testing simple measures 

During the lecture today we've learned about density and degrees. 

In `NetworkX` there are useful functions and methods that allow to calculate these measures. 

We use the `G.degree()` method for a list of all degrees of all the nodes. 

We use the  `nx.density(G)` function to calculate the density of the network. 

```python
print("The density is:", nx.density(G))
print("The degrees of my nodes are:", G.degree())
```

In order to get a degree of a selected node, e.g. Margot Robbie we specify this node in the degree method: 

```python
G.degree("Margot Robbie")
```
