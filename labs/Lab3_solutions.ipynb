Start by importing `networkx` and loading a graph from `data` folder. 

```python
import networkx as nx
G = nx.read_graphml('data/openflights_usa.graphml.gz')
```


### Exercise 1

Is there a direct flight between Indianapolis and Fairbanks, Alaska (FAI)? A direct flight is one with no intermediate stops.

```python
nx.has_path(G, "IND", "FAI")
```

```python
nr_of_links = nx.shortest_path_length(G, "IND", "FAI")
if nr_of_links == 1:
    print("Yes you can make it in one goe")
else:
    print("No, you will need to take", nr_of_links, "flights")
```

### Exercise 2

If I wanted to fly from Indianapolis to Fairbanks, Alaska what would be an itinerary with the fewest number of flights?

```python
nx.shortest_path(G, 'IND', 'FAI')
```

### Exercise 3

Is it possible to travel from any airport in the US to any other airport in the US, possibly using connecting flights? In other words, does there exist a path in the network between every possible pair of airports?

```python
nx.is_connected(G)
```

To check how many components they are in the network: 

```python
nx.number_connected_components(G)
```

To get a list of all nodes in each connected component: 

```python
list(nx.connected_components(G))
```

To check what are the names of the airports in the small components:

```python
# You could go one by one
G.nodes["AKB"]["name"]
```

```python
# but you can also use the min length and then loop through the elements: 

smallest_component = min(nx.connected_components(G), key=len)
for i in smallest_component: 
    print(G.nodes[i])
```
